```python
distancia = float(input("Ingrese la distancia entre las cámaras (en metros): "))
velocidad_max = float(input("Ingrese la velocidad máxima permitida (en km/h): "))
tiempo = float(input("Ingrese el tiempo que tardó el conductor en segundos: "))

# Mostrar los valores ingresados
print("Distancia:", distancia, "metros")
print("Velocidad máxima permitida:", velocidad_max, "km/h")
print("Tiempo:", tiempo, "segundos")

if distancia <= 0 or velocidad_max <= 0 or tiempo <= 0:
    print("ERROR")
else:
    velocidad_promedio = (distancia / tiempo) * 3.6  # Convertimos m/s a km/h

    if velocidad_promedio <= velocidad_max:
        print("Está bien")
    elif velocidad_promedio <= velocidad_max * 1.2:
        print("Multa")
    else:
        print("Multa y curso de capacitación")
```
